# Step 1 Removing Symbols and stop words from the articles

```python
import pandas as pd
import re
import nltk 
wn=nltk.WordNetLemmatizer()
nltk.download('stopwords')
nltk.download('wordnet')
import string
pd.set_option("display.max_colwidth",100)
stopwords=nltk.corpus.stopwords.words("english")
```

```python
SYMBOLS = '{}()[].,:;+-*/&|<>=~$1234567890'
Sym=list(SYMBOLS)
stopwords
```

# Removing stop words and symbols

```python
def clean_text(txt):
    for i in Sym:
        txt.replace(i,"")
    tokens=re.split("\W+",txt)
    txt="".join([c for c in txt if c not in string.punctuation])
    
    txt=[word for word in tokens if word not in stopwords]
    return txt
```

```python
df["content_without_symbols_and_stopWords"]=df["content"].apply(lambda x : clean_text(x))
```

# Applying Lemmatization

```python

def lemmatization(token_txt):
    text=[wn.lemmatize(word) for word in token_txt]
    return text
```

```python
df["lemmatized_content"]=df["content_without_symbols_and_stopWords"].apply(lambda x : lemmatization(x))
```

# ApplyinG Kmeans CLustring

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
```

#Converting limitized data list into string again

```python

for i in  range(0,len(df["lemmatized_content"])):
    p=df["lemmatized_content"][i]
    str1= " ".join([str(item) for item in p])
    df["lemmatized_content"][i]=str1
        
```

```python
do=df["lemmatized_content"].values.astype("U")
```

```python
vectorizer=TfidfVectorizer(stop_words='english')
features=vectorizer.fit_transform(do)
```

```python
k=7
model=KMeans(n_clusters=k,init='k-means++',max_iter=1000000000,n_init=1)
model.fit(features)
```

Assinging cluster numbers to articles

```python
df["cluster"]=model.labels_
```

# Printing Articles names along with cluster number they belong

```python
size=len(df["cluster"])
    
for i in range(0,k):
    
    print("Articles that belongs to cluster:",i)
    print("\n...............................................................................................\n")
    print("\n")
    for j in range(0,size):
        if df["cluster"][j]==i:
            print(df["title"][j])
    print("\n...............................................................................................\n")

```

# Featuers that Kmeans extracted and made cluster of articles according to those featureas

```python
print("Cluster centroids: \n")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()

for i in range(k):
    print("Cluster  %d features that kMeans found:" % i)
    for j in order_centroids[i, :10]:
        print (' %s' % terms[j])
    print('------------')
```
